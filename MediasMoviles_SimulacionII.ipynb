```python
import numpy as np
from statistics import mean,stdev,pstdev
from math import log
import pandas_datareader as pdr
import datetime
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
def calcular_medias_moviles(datos):
    medias_moviles = []
    k=10
    for t in range(k - 1, len(datos)):
        ventana = datos[t - (k - 1):t + 1]
        media = sum(ventana) / k
        medias_moviles.append(media)

    return medias_moviles
```

```python
data=yf.download("NKE.MX", start="2023-01-01", end="2023-06-01")

data.to_csv("datos.csv")
```

```python
plt.figure(figsize=(10,5))
plt.plot(data["Close"])
plt.title("NKE.MX STOCK CHART")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.show()
```

```python
media_nke=calcular_medias_moviles(data["Close"])
```

```python
plt.figure(figsize=(10,5))
plt.plot(media_nke)
plt.title("NKE.MX STOCK CHART")
plt.xlabel("Fecha")
plt.ylabel("PMedia")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
ax = plt.gca()

# Grfico de la serie de precios
ax.plot(data, color='blue', label='Precio')

# Grfico de la media mvil
ax.plot(media_nke, color='red', label='Media mvil')

# Ttulos y etiquetas de los ejes
plt.title("NKE.MX STOCK CHART")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.legend()

# Mostrar la grfica
plt.show()
```
