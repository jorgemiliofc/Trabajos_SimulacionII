```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
```

```python
def doce_u():
  a=0
  for i in range(12):
    u_i=rnd.random()
    a+=u_i
  return(a-6.0)
```

```python
lista=[]
for i in range (100000):
  lista.append(doce_u())
```

```python
plt.hist(lista,bins=500,color="green",rwidth=0.9)
plt.show()
```

```python
def aprox_u():
  u_i=rnd.random()
  x=((u_i)**0.135)-(1-u_i)**0.135
  return(x/0.1975)
```

```python
lista2=[]
```

```python
for i in range(10000):
  lista2.append(aprox_u())
```

```python
plt.hist(lista2,bins=500,color="red",rwidth=0.9)
plt.show()
```

```python
np.mean(lista2)
```

```python
np.var(lista2)
```
