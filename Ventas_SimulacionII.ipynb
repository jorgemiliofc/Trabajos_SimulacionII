```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def PRODUCTOS_VENDIDOS():
    ventas=0
    u = random.random()
    if 0.0 <= u < 0.2222:
        ventas = 100
    elif 0.2222 <= u < 0.4722:
        ventas = 200
    elif 0.4722 <= u < 0.5833:
        ventas = 300
    elif 0.5833 <= u < 0.8333:
        ventas = 400
    elif 0.8333 <= u <= 0.9444:
        ventas = 500
    elif 0.9444 <= u <= 1.00:
        ventas = 600
    return ventas
```

```python
def BEN(k): ##K = UNIDADES COMPRADAS
  VENTAS_T0=PRODUCTOS_VENDIDOS()
  COSTO_VENTA=1000
  COSTO_COMPRA=750
  INDEM_P_NOVENDIDO= 250

  if k>VENTAS_T0:
    ING_VENTAS=VENTAS_T0*COSTO_VENTA
    DEVUELTO=k-VENTAS_T0
    ING_DEVOLUCIONES=DEVUELTO*INDEM_P_NOVENDIDO
  elif k<=VENTAS_T0:
    ING_VENTAS=k*COSTO_VENTA
    DEVUELTO=0
    ING_DEVOLUCIONES=0

  C=k*COSTO_COMPRA
  B=ING_VENTAS+ING_DEVOLUCIONES-C

  #print(VENTAS_T0)
  #print(ING_VENTAS)
  #print(ING_DEVOLUCIONES)
  #print(C)
  #print(ING_VENTAS,"+", ING_DEVOLUCIONES, "-",C,"=",B)

  return(B)
```

```python
BEN(100)
```

```python
def experimento(n,k): #Hacemos el experimento n veces
    i = 0
    c=0
    Lista = []
    while i<=n:
        c+=BEN(k)
        i += 1
        p=c/i #probabilidad acumulativa
        Lista.append(p)
    return Lista
```

```python
exp=experimento(5000,400)
```

```python
def simulacion_ventas(n):
    List = []
    for i in range(n):
        List.append(np.mean(experimento(5000,400)))
    return List
```

```python
x=np.mean(experimento(5000,400))
print(x)
```

```python
rango = [100, 200, 300, 400, 500, 600]

plt.style.use('seaborn')
for i in rango:
    exp = experimento(5000, i)
    plt.plot(exp, linewidth=1.5, label=f'LOTE = {i}')
    plt.xlabel('ITERACIONES', fontsize=12)
    plt.ylabel('BENEFICIO', fontsize=12)
    plt.title('Beneficio', fontsize=14)
    plt.legend(loc='upper right', fontsize=10)
    plt.margins(0.05)
    plt.show()

    print("El beneficio converge a: ",np.mean(exp))
```
