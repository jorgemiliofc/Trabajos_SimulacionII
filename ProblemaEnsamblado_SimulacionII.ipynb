# **Problema ensamblado**

Tiempo de llegada de piezas a una estación de ensamblado sigue una distribucion exponencial, con una media

$\mu_{piezas}=2 \min$

Tiempo de duración de la tarea llevada acabo por el operario sigue una distribucion exponencial, con una media

$\mu_{tarea}= 1 \min$

Determine el tiempo medio de la permanencia de las piezas en dicho proceso




 *Notas:*

 **Simulación mediante eventos discretos**


 Para realizarla, damos seguimiento a ciertas variables:

1.   Variable de tiempo $t$: Se refiere al tiempo (simulado) que ha transcurrido
2.   Variables de conteo: Numero de veces que han ocurrido ciertos eventos hasta el momento t
3. Variable de estado del sistema: Describe el estado del sistema en el tiempo t

*Ross (pg. 78)*













```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#Paso 1: Sea t=s
s=0
t=s
n=100000
promedio_pieza=2
```

# Comprobación de la función exponencial

```python
#Veamos si la funcion si es exponencial

Lista=[]
for i in range(n):
  U=random.random()
  r=-(1/promedio_pieza)*np.log(U)
  Lista.append(r)
```

```python
plt.hist(Lista,bins=500,color="green",rwidth=0.9)
plt.show()
```

```python
lista=[]
for i in range(n):
  q=random.expovariate(promedio_pieza)
  lista.append(q)
```

```python
plt.hist(lista,bins=500,color="orange",rwidth=0.9)
plt.show()
```

# Solución

```python
T=1440 #(1 Dia)
Lista=[]
I=0
while t<T:
  U=random.random()
  t+= -(1/promedio_pieza)*np.log(U)
  I+= 1
  Lista.append(t)
```
